```python
import pandas as pd
import matplotlib.pyplot as plt
PATH = "ludnosc.csv"
df_ludnosc = pd.read_csv(PATH, sep = ';', decimal = ',')

df_ludnosc.pop("Unnamed: 8")
indexes = df_ludnosc[df_ludnosc.Rok == 1998].index
df_ludnosc.drop(indexes, inplace=True)
df_ludnosc
```

```python
region = df_ludnosc.Nazwa.unique()
rok = df_ludnosc.Rok.unique()
```

```python
df_ludnosc[(df_ludnosc.Kategoria == "Wydatki") & (df_ludnosc.Zmienna == "edukacja") & (df_ludnosc.Nazwa == "POLSKA")]
```

```python
df_ludnosc[(df_ludnosc.Kategoria == "Wydatki") & (df_ludnosc.Zmienna == "rekreacja i kultura") & (df_ludnosc.Nazwa == "POLSKA")]
```

```python
wydatki_kultura = df_ludnosc[(df_ludnosc.Kategoria == "Wydatki") & (df_ludnosc.Zmienna == "rekreacja i kultura") & (df_ludnosc.Nazwa == "POLSKA")].Wartosc.values
wydatki_edukacja = df_ludnosc[(df_ludnosc.Kategoria == "Wydatki") & (df_ludnosc.Zmienna == "edukacja") & (df_ludnosc.Nazwa == "POLSKA")].Wartosc.values
wydatki_combined = []
for i in range (0, len(wydatki_kultura)):
    wydatki_combined.append(wydatki_kultura[i]+wydatki_edukacja[i])
wydatki_combined
```

```python
wydatki_ogolem = df_ludnosc[(df_ludnosc.Kategoria == "Wydatki") & (df_ludnosc.Zmienna == "ogółem") & (df_ludnosc.Nazwa == "POLSKA")].Wartosc.values
```

```python
stosunek_edukacja_kultura_vs_ogolem = []
for i in range (0, len(wydatki_combined)):
    stosunek_edukacja_kultura_vs_ogolem.append(wydatki_combined[i]/wydatki_ogolem[i])
    stosunek_edukacja_kultura_vs_ogolem[i]*=100
    
stosunek_edukacja_kultura_vs_ogolem
```

```python
dochody_all = df_ludnosc[(df_ludnosc.Kategoria == "Dochody") & (df_ludnosc.Zmienna == "ogółem") & (df_ludnosc.Nazwa == "POLSKA")].Wartosc.values
```

```python
stosunek_edu_kult_vs_dochody = []
for i in range (0, len(wydatki_combined)):
    stosunek_edu_kult_vs_dochody.append(wydatki_combined[i]/dochody_all[i])
    stosunek_edu_kult_vs_dochody[i]*=100
    
stosunek_edu_kult_vs_dochody
```

```python
for i in range (0, len(region)):
    wydatki_edukacja = df_ludnosc[(df_ludnosc.Kategoria == "Wydatki") & (df_ludnosc.Zmienna == "edukacja") & (df_ludnosc.Nazwa == region[i])].Wartosc.values
    wydatki_kultura= df_ludnosc[(df_ludnosc.Kategoria == "Wydatki") & (df_ludnosc.Zmienna == "rekreacja i kultura") & (df_ludnosc.Nazwa == region[i])].Wartosc.values
    wydatki_combined = []
    for j in range (0, len(wydatki_kultura)):
        wydatki_combined.append(wydatki_kultura[j]+wydatki_edukacja[j])
    stosunek_edu_kult_vs_dochody = []
    for k in range (0, len(wydatki_combined)):
        stosunek_edu_kult_vs_dochody.append(wydatki_combined[k]/dochody_all[k])
        stosunek_edu_kult_vs_dochody[k]*=100
    plt.scatter(rok, stosunek_edu_kult_vs_dochody)
    plt.xlabel('Rok')
    plt.ylabel('wydatki kultura [%]')
    plt.title('Udział wydatków na kulturę i edukację w stosunku do dochodów na przestrzeni lat na terenie: %r' %(region[i]))
    plt.show()
```
